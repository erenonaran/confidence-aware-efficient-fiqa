```python
import numpy as np
np.bool = bool

from face_image_quality import SER_FIQ
import cv2
import glob
import csv


ser_fiq = SER_FIQ(gpu=0)

model = "rf"
operation = "aligned"
csv_file = f"/content/ijbb_{model}_{operation}_serfiq.csv"

with open(csv_file, mode='w', newline='') as file:
  csv_writer = csv.writer(file)
  csv_writer.writerow(["Image","Score"])

  for i,image_path in enumerate(glob.glob(f"/content/ijbb_{model}_{operation}/*.jpg")):
    cropped_img = cv2.imread(image_path)
    resized_image = cv2.resize(cropped_img, (112, 112))
    score = ser_fiq.get_score(np.transpose(resized_image,(2,0,1)), T=100)
    csv_writer.writerow([image_path.split("/")[-1],score])
    print(f"SER-FIQ quality score of image {i}-{image_path} is", score)
```

```python
import numpy as np
np.bool = bool

from face_image_quality import SER_FIQ
import cv2
import glob
import csv


ser_fiq = SER_FIQ(gpu=0)

for d,c in [(d,c) for d in range(1,4) for c in range(1,6)]:
  for (op, model) in [(op, model) for op in ["cropped","aligned"] for model in ["mtcnn","rf"]]:
    scface = f'scface_d{d}_c{c}_{model}_{op}'
    csv_file = f"/content/{scface}_serfiq.csv"

    with open(csv_file, mode='w', newline='') as file:
      csv_writer = csv.writer(file)
      csv_writer.writerow(["Image","Score"])

      for i,image_path in enumerate(glob.glob(f"/content/{scface}/*.jpg")):
        cropped_img = cv2.imread(image_path)
        resized_image = cv2.resize(cropped_img, (112, 112))
        score = ser_fiq.get_score(np.transpose(resized_image,(2,0,1)), T=100)
        csv_writer.writerow([image_path.split("/")[-1],score])
        print(f"SER-FIQ quality score of image {i}-{image_path} is", score)
```
